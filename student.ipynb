# Data cleaning Strategy

After going through all the data that, I chose to either drop the missing values or replace the null values with 0 for the following reasons:

1. we will have sufficient data to work with when making the analysis even after dropping or replacing null values with zero in these rows.

2. The data in Box office Movies gross is not as much the data in IMDB movie title it is very little and missing values make up approximately 40% of the data therefore in order for data to be continous I chose to replace all the null values in the columns with missing data in  with 0

3. The column Original title will be dropped so that we can focus on the primary title since primary title is the translated version of it and we won''t be needing two titles

4. We can't use the most occuring data to fill in data on any of these missing data rows for example the column Genre I can't deduce that this movie's genre is comedy based on the title or the most occuring genres. It is the same on the other columns such as runtime_minutes, studio , domestic_gross and foreign_gross . 

5. The Data on this rows cannot be replaced or predicted as they are unique.

# Combining the data frames for Movie title, ratings and gross

First lets set index on Movie titles and ratings df then we can combine Titles_df and Ratings_df using Inner join so that we can get rid of the datas that won't match in both data frames

```python
#set index tconst on movie title data frame
Titles_df.set_index('tconst', inplace=True)
Titles_df.head()
```

```python
#set index tconst on movie ratings data frame
Ratings_df.set_index('tconst', inplace=True)
Ratings_df.head()
```

###### Great job! now lets combine these two data frames.

```python
Titles_with_ratings_df = Titles_df.join(Ratings_df, how='inner')
Titles_with_ratings_df 
```

```python
#As you can see above we've lost one column "tconst" because we set it as index now lets reverse it
Titles_with_ratings_df.reset_index(inplace=True)
Titles_with_ratings_df.info()
```

#### Wonderful! now lets combine our new data frame with the movie gross
1. But first have you noticed that in Box movie gross the title of the movie column is saved as 'title' and in titles_with_ratings_df it is saved as Primary_title and Original_title? ok now lets hope and pray that Primary_title is same as title.

2. We will first change column name 'title'in Box office movies gross to 'Primary_title' then we can make it the primary key and we can combine these two data frames

```python
#lets change the column name 'title' in Movie_gross_df to 'primary_title'
Movie_gross_df = Movie_gross_df.rename({'title': 'primary_title'}, axis=1) 
Movie_gross_df.head()
```

```python
#great! now lets set 'primary_title' as index on both Movie_gross_df and Titles_with_ratings_df
Movie_gross_df.set_index('primary_title', inplace=True)
Movie_gross_df.head()
```

```python
Titles_with_ratings_df.set_index('primary_title', inplace=True)
Titles_with_ratings_df.head()
```

```python
#great! now lets combine the two data frames 
Movie_details_df = Titles_with_ratings_df.join(Movie_gross_df, how='inner')
Movie_details_df
```

```python
Movie_details_df.info()
```

```python
#As you can see above we've lost one column "primary_title" because we set it as index now lets reverse it
Movie_details_df.reset_index(inplace=True)
Movie_details_df
```

```python
#now lets check on some duplicate values in the column Primary title
Movie_details_df['primary_title'].duplicated().sum()
```

```python
#lets drop the duplicated columns
Movie_details_df.drop_duplicates(subset="primary_title", keep= 'first', inplace=True)
Movie_details_df
```

```python
Movie_details_df['primary_title'].duplicated().sum()
```

```python
Movie_details_df.describe()
```

```python
Movie_details_df.head(5)
```

```python
#lets create a new column 'gross_income' by getting the total of domestic_gross and foreign_gross
Movie_details_df = Movie_details_df.eval("gross_income = domestic_gross + foreign_gross")
Movie_details_df.head(5)
```

```python
#lets find the studios that produced the movies with the highest gross income
Highest_income = Movie_details_df.groupby(['studio'])['gross_income'].sum().nlargest(10)
Highest_income
```

```python
#lets plot the studios that produced movies which brought in the highest gross income

fig = plt.figure(figsize=(17, 6))
ax = fig.add_subplot(111)

width = 0.5

Highest_income.plot(kind='bar', color= 'pink', ax=ax, width=width, position=1)
plt.xlabel('Studio', fontsize = 16)
plt.ylabel('Frequency', fontsize = 16)
ax.set_title("Top ten Profitable Studios", fontsize = 16)

ax.legend()
plt.show()
```

```python
#lets find the top 10 highest rated movies 
top_rated_movies = Movie_details_df.groupby(['primary_title'])['averagerating'].sum().nlargest(10)
top_rated_movies
```

```python
#lets plot the top 10 highest rated movies against their ratings

fig = plt.figure(figsize=(17, 6))
ax = fig.add_subplot(111)

width = 0.5

top_rated_movies.plot(kind='barh', color= 'pink')
plt.xlabel('Ratings ', fontsize = 16)
plt.ylabel('Movie', fontsize = 16)
ax.set_title("Top 10 highest rated movies", fontsize = 16)

ax.legend()
plt.show()
```

```python
top_100_genres = Movie_details_df.groupby(['genres'])['averagerating'].sum().nlargest(100)
top_100_genres
```

```python
#now that we can see we are unable to plot the data in genres because of the mixed data lets try and flatten it
Split_genres = Movie_details_df.copy()
Split_genres['genres'] = Split_genres['genres'].str.split(',').copy()

Split_genres = Split_genres.explode('genres')
Split_genres.head(10)
```

```python
top_10_genres = Movie_details_df.groupby(['genres'])['gross_income'].sum().nlargest(10)
top_10_genres
```

```python
#lets plot the top 10 highest genres against their gross income

fig = plt.figure(figsize=(17, 6))
ax = fig.add_subplot(111)

width = 0.5

top_10_genres.plot(kind='barh', color= 'green')
plt.xlabel('Income', fontsize = 16)
plt.ylabel("Genre", fontsize = 16)
ax.set_title("Top 10 most profitable genres", fontsize = 16)

ax.legend()
plt.show()
```
